```python
from tqdm import tqdm



num = 1
n = 2
for idx, i in tqdm(enumerate(gu_list[1:])):

            
    # 스벅 화면 스크롤 내리기
    driver.execute_script('window.scrollTo(0, 60);')
    time.sleep(3)
    
    
    # 스타벅스 페이지는 웹페이지가 바뀌는 게 아니라 지역검색으로 들어가서 다시 구를 눌러줘야하는 형태
    # 지역검색 클릭하기
    local_search_btn = driver.find_element(By.XPATH, '//*[@id="container"]/div/form/fieldset/div/section/article[1]/article/header[2]/h3/a')
    local_search_btn.click()
    time.sleep(3)

    # 시도를 서울로 선택하여 클릭
    sido_btn = driver.find_element(By.XPATH, '//*[@id="container"]/div/form/fieldset/div/section/article[1]/article/article[2]/div[1]/div[2]/ul/li[1]/a')
    sido_btn.click()
    time.sleep(3)
    
    print(idx)
    if idx >= 20:
        # 스벅 내에 사이드 창 내리는 코드 
        side_window = driver.find_element(By.XPATH, '//*[@id="mCSB_2_dragger_vertical"]/div')
        actions = ActionChains(driver)
        actions.move_to_element(side_window).click_and_hold().move_by_offset(0, 70).release().perform()
        time.sleep(3)

    # 구를 선택
        gu_btn = driver.find_element(By.XPATH, '//*[@id="mCSB_2_container"]/ul/li[{}]/a'.format(n))
                                            
        gu_btn.click()
        time.sleep(3)

    else:
        gu_btn = driver.find_element(By.XPATH, '//*[@id="mCSB_2_container"]/ul/li[{}]/a'.format(n))                               
        gu_btn.click()
        time.sleep(3)
    # 현재 페이지의 html 코드 가져오기
    req = driver.page_source    

    # BeautifulSoup을 이용
    soup = BeautifulSoup(req, 'html.parser')

    # 매장 정보 가져오기
    store = soup.find_all('li', 'quickResultLstCon')[10:]




    store_num = 0
    for j in tqdm(store):
            
       
        # 스벅 매장명 가져오기
        name = store[store_num].find('strong').text.strip()

        # 스벅 주소 가져오기
        address = store[store_num].find('p').text


        # 스벅 구 가져오기
        gu = store[store_num].find('p').text.split()[1]

        # 위도, 경도 가져오기
        lat = store[store_num]['data-lat']
        lng = store[store_num]['data-long']
        
        a = [num, 1, name, gu, address, lat, lng]
        cursor.execute('insert into COFFEE_STORE values (%s, %s, %s, %s, %s, %s, %s)', tuple(a))
        conn.commit()
        
        store_num += 1
        num += 1

        
    n += 1 # 다음 구 입력
```

```python
# 이디야 매장 주요 분포 지역 (매장수가 많은 상위 5개 구이름, 매장 개수 출력)
cursor.execute('select gu_name, count(*) as 매장개수 from COFFEE_STORE where brand = 2 group by gu_name order by 매장개수 desc limit 5')
cursor.fetchall()
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 브랜드이름, 매장 개수 출력)
cursor.execute('select gu_name, brand, count(*) as 매장개수 from COFFEE_STORE group by gu_name, brand order by gu_name')
cursor.fetchall()
```

```python
# 구별 브랜드 각각의 매장 개수 조회 (구이름, 스타벅스 매장 개수, 이디야 매장 개수 출력)
cursor.execute('select 스타벅스.gu_name, 스타벅스.매장개수 as 스타벅스, 이디야.매장개수 as 이디야 from (select gu_name, count(*) as 매장개수 from COFFEE_STORE where brand = 1 group by gu_name order by gu_name) 스타벅스 left join (select gu_name, count(*) as 매장개수 from COFFEE_STORE where brand = 2 group by gu_name order by gu_name) 이디야 on 스타벅스.gu_name = 이디야.gu_name')
cursor.fetchall()
```
